```python
from pathlib import Path
import os
import re
import apsw
import pandas as pd
import time

```

```python
cwd = Path.cwd().parent.parent
print(cwd)
```

```python
dbs = [
    os.path.join(cwd, 'db', 'db01', 'NetCollector.sqlite'),
    os.path.join(cwd, 'db', 'db02', 'NetCollector.sqlite'),
]
```

```python
from time import time
from sqlalchemy import create_engine
start = time()
df1 = None
df2 = None
for idx, db_path in enumerate(dbs):
    engine = create_engine(f'sqlite:////{db_path}')

    print(engine)

    sql = """

    select d.direction,
           d.device_type
    from deviceLog d
    where d.device_type is not null

    """

    if idx == 0:
        df1 = pd.read_sql_query(sql, engine)
    elif idx == 1:
        df2 = pd.read_sql_query(sql, engine)

end = time()
print(f'[+] Finished processing in {(end-start):.2f} seconds')
```

```python
df = pd.concat([df1, df2], axis='rows', ignore_index=True)
df.describe()
```

```python
df['direction'].unique()
```

```python
def title(s):
    val = s[0].upper() + s[1:]
    return f'{val} Packet Count by IoT Device Type'


import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
exclude_list = []
ds = [x for x in sorted(df['direction'].unique()) if x not in exclude_list]
my_colors = 'rgbkymc'

for idx, item in enumerate(ds, 1):
    arr = []
    for k, v in df[((df['direction']==item) & (df['device_type']!='Other')) ].groupby('device_type').count()['direction'].iteritems():
        arr.append([k, v])
    arr = list(reversed(sorted(arr, key=lambda x: x[1])))
    x = [p[0] for p in arr]
    y = [p[1] for p in arr]
    plt.rcParams["figure.figsize"] = (10,15)
    plt.rcParams["figure.dpi"] = 150
    plt.suptitle('Traffic Stats by Device Type', fontsize=16, fontweight='bold')
    plt.subplot(3,1,idx)
    plt.subplots_adjust(wspace = 0.5, hspace = 0.8) 
    plt.bar(x, y, color=my_colors)
    plt.yscale('log')
    plt.title(title(item), fontsize=14, fontweight='bold')
    plt.xticks(rotation=90, fontsize=10)
    plt.yticks(fontsize=10)
    plt.xlabel('Device Type', fontsize=12, fontweight='bold')
    plt.ylabel('Packet Count (Log Scale)', fontsize=12, fontweight='bold')   
plt.show()
```
