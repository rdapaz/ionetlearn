```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = """
Other|13632|41.6%
Yamaha Amp|11888|36.2%
iZone Controller|3142|9.6%
Samsung Phone|2632|8.0%
Amazon Echo|563|1.7%
Light Bulbs LiFX Smart Bulb|183|0.6%
Samsung SmartCam|146|0.4%
Withings Aura smart sleep sensor|136|0.4%
Samsung Galaxy Tab|135|0.4%
Sonos Speaker|109|0.3%
Other values (14)|233|0.7%
""".splitlines()

data = [x.split('|') for x in data if len(x) > 0]
new_data = []
for device, count, perc in data:
    count = int(count)
    perc = float(perc[:-1])/100.0
    new_data.append([device, count, perc])
```

```python
headings = ['Device Type', 'Count', 'Perc']
df = pd.DataFrame(data=new_data, columns=headings)
```

```python
sns.set(style="whitegrid")
```

```python
a4_dims = (16.53, 11.7)
fig, ax = plt.subplots(figsize=a4_dims)
ax = sns.barplot(x="Count", y="Device Type", data=df, palette="Greys_d")
fig.savefig('barplot.png')
```
